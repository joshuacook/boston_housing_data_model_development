# Visualize Features

```python
ames_train_df = pd.read_pickle('data/train_cblock_df.p')
```

```python
X = ames_train_df.drop('SalePrice', axis=1)
y = ames_train_df['SalePrice']
```

```python
numerical_features = set()
for col in X:
    if X[col].dtype == np.dtype('float'):
        numerical_features.add(col)

all_features = set(X.columns)
categorical_features = all_features - numerical_features 
numerical_features = list(numerical_features)
categorical_features = list(categorical_features)
```

```python
len(numerical_features), len(categorical_features)
```

```python
X_num = X[numerical_features]
X_cat = X[categorical_features]
```

```python
fig = plt.figure(figsize=(16,20))
for i, col in enumerate(X_num):
    fig.add_subplot(5,4,i+1)
    sns.regplot(col, y, data=ames_train_df)
```
